```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
t = np.arange(-5, 5, 0.0001)
k = 4/pi
m = pi/2

k1 = k * np.cos(m*t)
k2 = (-k/3) * np.cos(3*m*t)
k3 = (k/5) * np.cos(5*m*t)
k4 = (-k/7) * np.cos(7*m*t)
k5 = (k/9) * np.cos(9*m*t)

amp = k1+k2+k3+k4+k5
```

```python
plt.plot(t, amp)
```
