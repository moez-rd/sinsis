```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter simulasi
fs = 1000  # Frekuensi sampling (Hz)
T = 1/fs   # Periode sampling
duration = 1  # Durasi sinyal (detik)
t = np.arange(0, duration, T)  # Vektor waktu

# Sinyal analog (contoh: sinusoidal)
f_signal = 5  # Frekuensi sinyal analog (Hz)
analog_signal = np.sin(2 * np.pi * f_signal * t)

plt.figure(figsize = (16,16)) # set the size of figure


# Plot sinyal analog
plt.subplot(4, 1, 1)
plt.plot(t, analog_signal)
plt.title('Sinyal Analog')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)

# 1. Sampling
sampled_signal = analog_signal[::10]  # Mengambil setiap 10 sampel
t_sampled = t[::10]

# Plot sinyal setelah sampling
plt.subplot(4, 1, 2)
plt.stem(t_sampled, sampled_signal, markerfmt='ro', basefmt='r-', linefmt='r-')
plt.title('Sinyal Setelah Sampling')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)

# 2. Quantisasi
quantized_signal = np.round(sampled_signal)

# Plot sinyal setelah quantisasi
plt.subplot(4, 1, 3)
plt.stem(t_sampled, quantized_signal, markerfmt='go', basefmt='g-', linefmt='g-')
plt.title('Sinyal Setelah Quantisasi')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo (Hasil Quantisasi)')
plt.grid(True)

# 3. Encoding
binary_signal = np.unpackbits(quantized_signal.astype(np.uint8))

# Plot sinyal setelah encoding
plt.subplot(4, 1, 4)
plt.stem(range(len(binary_signal)), binary_signal, markerfmt='bo', basefmt='b-', linefmt='b-')
plt.title('Sinyal Setelah Encoding')
plt.xlabel('Bit')
plt.ylabel('Nilai Biner')
plt.grid(True)

plt.tight_layout()
plt.show()

# 4. Decoding
decoded_signal = np.packbits(binary_signal).astype(np.float64)

# 5. Rekonstruksi Sinyal
reconstructed_signal = np.repeat(decoded_signal, 10)  # Menyusun ulang sinyal setelah decoding

# Plot sinyal setelah rekonstruksi
plt.figure()
plt.plot(t, reconstructed_signal)
plt.title('Sinyal Setelah Rekonstruksi')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.show()
```
