```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
frequency_sampling, audio_signal = wavfile.read("Tonmoy 07.wav")
print(len(audio_signal))
audio_signal = audio_signal[:50]
t = 1000 * np.arange(0, len(audio_signal), 1) / float(frequency_sampling)
x = audio_signal
```

```python
plt.plot(t, x)
plt.title('Analog Signal')
plt.xlabel('t')
plt.ylabel('amplitude')
plt.show()
```

```python
n = t

markerline, stemlines, baseline = plt.stem(n, x)
plt.setp(stemlines)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.show()
```

```python
plt.title('Quantization')

plt.plot(t, audio_signal)
markerline, stemlines, baseline=plt.stem(n, x)
plt.setp(stemlines)
plt.xlabel('n')
plt.ylabel('Range of Quantizer')

plt.axhline(y = -10, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -20, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -30, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -40, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -50, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -60, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -70, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -80, xmin = 0, xmax = 10, color = 'r')
```

```python
plt.title('Quantized Signal')
xq = np.around(x, 1)
markerline, stemlines, baseline = plt.stem(n, xq)
plt.setp(stemlines)
plt.xlabel('n')
plt.ylabel('Range of Quantizer')

plt.axhline(y = -10, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -20, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -30, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -40, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -50, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -60, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -70, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = -80, xmin = 0, xmax = 10, color = 'r')
```
