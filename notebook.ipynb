```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
frequency_sampling, audio_signal = wavfile.read("Tonmoy 07.wav")
print(len(audio_signal))
audio_signal = audio_signal[:50]
t = 1000 * np.arange(0, len(audio_signal), 1) / float(frequency_sampling)
x = -min(audio_signal) + audio_signal
```

```python
plt.plot(t, x)
plt.title('Analog Signal')
plt.xlabel('t')
plt.ylabel('amplitude')
plt.show()
```

```python
x_scaled = x / max(x)
```

```python
plt.plot(t, x_scaled)
plt.title('Analog Signal')
plt.xlabel('t')
plt.ylabel('amplitude')
plt.show()
```

```python
n = t

markerline, stemlines, baseline = plt.stem(n, x_scaled)
plt.setp(stemlines)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.show()
```

```python
plt.title('Quantization')

plt.plot(t, x_scaled)
markerline, stemlines, baseline=plt.stem(n, x_scaled)
plt.setp(stemlines)
plt.xlabel('n')
plt.ylabel('Range of Quantizer')

plt.axhline(y = 0.1, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.2, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.3, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.4, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.5, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.6, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.7, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.8, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.9, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 1.0, xmin = 0, xmax = 10, color = 'r')
```

```python
plt.title('Quantized Signal')
x_quantized = np.around(x_scaled, 1)
markerline, stemlines, baseline = plt.stem(n, x_quantized)
plt.setp(stemlines)
plt.xlabel('n')
plt.ylabel('Range of Quantizer')

plt.axhline(y = 0.1, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.2, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.3, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.4, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.5, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.6, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.7, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.8, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 0.9, xmin = 0, xmax = 10, color = 'r')
plt.axhline(y = 1.0, xmin = 0, xmax = 10, color = 'r')
```

```python
x_encoded = []

for x in x_quantized:
    x_encoded.append(np.binary_repr(int(x*10), 4))
    
x_encoded = np.array(x_encoded)
x_encoded
```

```python
x_decoded = []

for x in x_encoded:
    x_decoded.append(int(x, 2))
    
x_decoded = np.array(x_decoded) / 10
```

```python
plt.title('Decoded Signal')
markerline, stemlines, baseline = plt.stem(n, x_decoded)
plt.setp(stemlines)
plt.xlabel('n')
plt.ylabel('Range of Quantizer')
```

```python
plt.title('Reconstucted Signal')
plt.plot(n, x_decoded)
plt.xlabel('n')
plt.ylabel('Range of Quantizer')
```
