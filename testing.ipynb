```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-12, 12, 1)
y1 = 1 + np.sin(x)
```

```python
print(x)
print(y1)
```

```python
plt.stem(x, y1)
plt.xlabel('independent variable')
plt.ylabel('dependent variable')
plt.title('simple graph')
plt.grid()
plt.show()
```
